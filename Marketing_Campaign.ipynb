# Repeated Stratified K-Fold Cross-Validation 
Repeated Stratified K-Fold Cross-Validation Example
This project demonstrates the use of RepeatedStratifiedKFold from the scikit-learn library for evaluating a model's performance using cross-validation. This method is particularly useful for datasets with imbalanced classes and provides a more stable estimate of model accuracy by repeating the stratified cross-validation multiple times.

Overview
RepeatedStratifiedKFold extends the StratifiedKFold approach by repeating the stratified fold process multiple times, which helps in achieving a more generalized and robust model evaluation. It ensures that each class is evenly represented in each fold and reduces variance associated with a single trial of cross-validation.

Python 3.8+
scikit-learn

from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

# Load dataset
data = load_iris()
X, y = data.data, data.target

model = SVC()

# Setup cross-validation
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=10, random_state=42)

# Evaluate model

print("Mean Accuracy: %0.2f" % scores.mean())
This script loads the Iris dataset, initializes an SVM classifier, and evaluates it using repeated stratified K-fold cross-validation. The results display the mean accuracy across all folds and repeats.

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
# Step 3: Scale the data (after resampling)
````

````python

# List of classifiers
classifiers = [
    ('RFC', RandomForestClassifier(random_state=42)),
    ('GBC', GradientBoostingClassifier(random_state=42)),
    ('KNN', KNeighborsClassifier()),
    ('SVC', SVC(random_state=42)),
    ('LR',  LogisticRegression()),
    ('DTC', DecisionTreeClassifier(random_state=42)),
    ('KMeans', KMeans())
]

# List of scalers
scalers = [
    ('SS', StandardScaler()),
    ('MMS', MinMaxScaler())
]

# List of resampling methods
samplers = [
    ('RUS', RandomUnderSampler(random_state=42)),
    ('ROS', RandomOverSampler(random_state=42)),
    ('SMOTE', SMOTE(random_state=42)),
    ('SMOTEENN', SMOTEENN(random_state=42)),
    ('ADASYN' , ADASYN(random_state=42))
]
scoring_metrics = {
    'accuracy': 'accuracy',
    'precision': 'precision_weighted',
    'recall': 'recall_weighted',
    'f1_score': 'f1_weighted'
}
````

````python

# Evaluating each combination of sampler, scaler, and classifier
results = []

# Run just the classifier
for classifier_name, classifier in classifiers:
    if classifier_name in ['LR', 'SVC', 'KNN']:  # Example: only use PCA with certain classifiers
        pipeline = make_imbalance_pipeline(PCA(n_components=8), classifier)  # Choose a reasonable number of components
    else:
        pipeline = make_imbalance_pipeline(classifier)
    cv = RepeatedStratifiedKFold(n_splits=3, n_repeats=10, random_state=42)
    scores = cross_validate(pipeline, X_train, y_train, scoring=scoring_metrics, cv=cv, n_jobs=-1)
    # Average each metric across folds
    mean_scores = {metric: np.mean(scores[f'test_{metric}']) for metric in scoring_metrics}
    results.append(('NA', 'NA', classifier_name, mean_scores['accuracy'],
                    mean_scores['precision'], mean_scores['recall'], mean_scores['f1_score']))
    
# Run Scaler and Classifier
for scaler_name, scaler in scalers:
    for classifier_name, classifier in classifiers:
        if classifier_name in ['LR', 'SVC', 'KNN']:  # Example: only use PCA with certain classifiers
            pipeline = make_imbalance_pipeline(scaler, PCA(n_components=8), classifier)
        else:
            pipeline = make_imbalance_pipeline(scaler, classifier)        
        cv = RepeatedStratifiedKFold(n_splits=3, n_repeats=10, random_state=42)
        scores = cross_validate(pipeline, X_train, y_train, scoring=scoring_metrics, cv=cv, n_jobs=-1)
        # Average each metric across folds
        mean_scores = {metric: np.mean(scores[f'test_{metric}']) for metric in scoring_metrics}
        results.append(('NA', scaler_name, classifier_name, mean_scores['accuracy'],
                        mean_scores['precision'], mean_scores['recall'], mean_scores['f1_score']))

# Run Sampler and Classifier
for sampler_name, sampler in samplers:
    for classifier_name, classifier in classifiers:
        if classifier_name in ['LR', 'SVC', 'KNN']:  # Example: only use PCA with certain classifiers
            pipeline = make_imbalance_pipeline(sampler, PCA(n_components=8), classifier)
        else:
            pipeline = make_imbalance_pipeline(sampler, classifier)
        cv = RepeatedStratifiedKFold(n_splits=3, n_repeats=10, random_state=42)
        scores = cross_validate(pipeline, X_train, y_train, scoring=scoring_metrics, cv=cv, n_jobs=-1)
        # Average each metric across folds
        mean_scores = {metric: np.mean(scores[f'test_{metric}']) for metric in scoring_metrics}
        results.append((sampler_name, 'NA', classifier_name, mean_scores['accuracy'],
                        mean_scores['precision'], mean_scores['recall'], mean_scores['f1_score']))

# Run Sampler, Scaler and Classifier
for sampler_name, sampler in samplers:
    for scaler_name, scaler in scalers:
        for classifier_name, classifier in classifiers:
            if classifier_name in ['LR', 'SVC', 'KNN']:  # Example: only use PCA with certain classifiers
                pipeline = make_imbalance_pipeline(scaler, sampler, PCA(n_components=8), classifier)
            else:
                pipeline = make_imbalance_pipeline(scaler, sampler, classifier)
            cv = RepeatedStratifiedKFold(n_splits=3, n_repeats=10, random_state=42)
            scores = cross_validate(pipeline, X_train, y_train, scoring=scoring_metrics, cv=cv, n_jobs=-1)
            # Average each metric across folds
            mean_scores = {metric: np.mean(scores[f'test_{metric}']) for metric in scoring_metrics}
            results.append((sampler_name, scaler_name, classifier_name, mean_scores['accuracy'],
                            mean_scores['precision'], mean_scores['recall'], mean_scores['f1_score']))

````

````python
# Convert results to DataFrame for easy viewing
results_df = pd.DataFrame(results, columns=['Sampler', 'Scaler', 'Classifier', 'Accuracy', 'Precision', 'Recall', 'F1-Score'])
sorted_results_df = results_df.sort_values(by=['Recall','Precision','F1-Score','Accuracy'], ascending=False)
print("The top 10 models by are:\n")
print(sorted_results_df.head(20).to_string(index=False))
````
